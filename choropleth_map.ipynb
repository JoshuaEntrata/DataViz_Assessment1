```python
pycountry.countries.search_fuzzy('Bolivia')
```

```python
def get_alpha_3(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except:
        return 'NONE'
    
merged_df['iso_alpha3'] = merged_df['area'].apply(get_alpha_3)
merged_df[merged_df['iso_alpha3'] == 'NONE'][['area', 'iso_alpha3']].drop_duplicates()
```

```python
manual_iso_map = {
    'Bolivia (Plurinational State of)': 'BOL',
    'China, Hong Kong SAR': 'HKG',
    'China, Macao SAR': 'MAC',
    'China, mainland': 'CHN',
    'China, Taiwan Province of': 'TWN',
    "CÃƒÂ´te d'Ivoire": 'CIV',
    'Iran (Islamic Republic of)': 'IRN',
    'Republic of Korea': 'KOR',
    'Turkey': 'TUR',
    'Venezuela (Bolivarian Republic of)': 'VEN',
    'Timor-Leste': 'TLS'
}

merged_df['iso_alpha3'] = merged_df['area'].map(manual_iso_map).fillna(merged_df['iso_alpha3'])

merged_df[['area', 'iso_alpha3']].drop_duplicates()
```

```python
merged_df[merged_df['iso_alpha3'] == 'NONE'][['area', 'iso_alpha3']].drop_duplicates()
```

```python
fig = px.choropleth(
    merged_df,
    locations='iso_alpha3',  
    color='value_per_capita', 
    hover_name='area', 
    color_continuous_scale=px.colors.sequential.Plasma,  
    projection='natural earth', 
    title='Average Production by Country'
)

fig.show()
```
