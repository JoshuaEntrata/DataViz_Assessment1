```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import warnings
```

## `Part 2: Data Exploration`


### **Descriptive Statistics**


#### Summarize Statistics


```python
# Generate summary statistics 
merged_df.describe()
```

```python
# Plot to check the outliers in the average production and value per capita

plt.figure(figsize=(10, 6))
sns.boxplot(x=merged_df['average_production'])
plt.title('Box Plot of Average Production')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=merged_df['value_per_capita'])
plt.title('Box Plot of Value per Capita')
plt.show()
```

#### Group-Based Statistics


```python
merged_df.groupby('area').agg({
    'value': ['mean', 'min', 'max', 'median', 'std']
})
```

```python
merged_df.groupby('area').agg({
    'average_production': ['mean', 'min', 'max', 'median', 'std'],
})
```

```python
merged_df.groupby('area').agg({
    'value_per_capita': ['mean', 'min', 'max', 'median', 'std']
})
```

```python
average_prod_by_area = merged_df.groupby('area')['average_production'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12, 24))
sns.barplot(
    x='average_production', 
    y='area', 
    data=average_prod_by_area,
)
plt.title('Average Production by Area')
plt.xlabel('Average Production')
plt.ylabel('Area')
plt.show()
```

```python
vpc_by_area = merged_df.groupby('area')['value_per_capita'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12, 24))
sns.barplot(
    x='value_per_capita', 
    y='area', 
    data=vpc_by_area,
)
plt.title('Value per Capita by Area')
plt.xlabel('Value per Capita')
plt.ylabel('Area')
plt.show()
```

```python
years_by_area = merged_df.groupby('area')['years_existing'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12, 24))
sns.barplot(
    x='years_existing', 
    y='area', 
    data=years_by_area,
)
plt.title('Years Existing by Area')
plt.xlabel('Years Existing')
plt.ylabel('Area')
plt.show()
```

### **Correlation Analysis**
